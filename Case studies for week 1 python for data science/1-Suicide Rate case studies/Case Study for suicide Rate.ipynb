# First import dataset and all libraries

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
masterdf = pd.read_csv('master.csv')
masterdf.describe()
```

# 1. Is the suicide rate more prominent in some age categories than others

```python
plt.figure(figsize=(10,5))
ageVsSuicideNum = sns.barplot(x='age', y='suicides_no', data=masterdf, palette='muted')
```

# In the following graph we can see 35-54 years are the more prominent age categories

# 2. Which countries have the most and the least number of suicides?

```python
masterdf.groupby(by=['country'])['suicides_no'].sum().reset_index().sort_values(['suicides_no']).tail(10).plot(x='country',
                                                                                                           y='suicides_no',
                                                                                                           kind='bar',
                                                                                                           figsize=(15,5))
plt.show()
```

# As we can see from this graph, the highest amount of suicides numerically is in the Russian Federation

```python
masterdf.groupby(by=['country'])['suicides_no'].sum().reset_index().sort_values(['suicides_no']).head(10).plot(x='country',
                                                                                                           y='suicides_no',
                                                                                                           kind='bar',
                                                                                                           figsize=(15,5))
plt.show()
```

# And as we can see from this graph the lowest number of suicides numerically is in Saint Kitts and Nevis and in the Dominica

# 3. What is the effect of the population on suicide rates?

```python
figure = plt.figure(figsize=(15,5))

populationVsSuicideRates = sns.scatterplot(x=masterdf['population'],y='suicides_no', data=masterdf, size = "suicides_no") 
```

# As the population number gets higher, the suicide numbers also go up with an anoumalous spike in the 2 millions of the population.

 # 4. What is the effect of the GDP of a country on suicide rates?

```python
figure = plt.figure(figsize=(15,7))
sns.scatterplot(x='gdp_per_capita ($)', y='suicides/100k pop', data=masterdf)
plt.show()
```

# As the GDP increases the suicide rates decrease. Which means that apparent to this graph the lower the GDP per capita is the greater the suicide rate is.

# 5. What is the trend of suicide rates across all the years?

```python
masterdf.groupby(by=['year'])['suicides/100k pop'].sum().reset_index().sort_values(['year']).plot(x='year',
                                                                                                           y='suicides/100k pop',
                                                                                                           kind='bar',
                                                                                                           figsize=(15,5))
```

# The biggest trend appeared from the years 1995-2003 with 1995 being the peak of suicide numbers thereafter, it slowly decreased but still remained very much increased from the late 80's and early 90's. We can also see an anomaly in 2016 as they where exponentialy lower 

# 6. Is there a difference between the suicide rates of men and women?

```python
plt.figure(figsize=(10,5))
sns.barplot(data=masterdf,x='sex',y='suicides/100k pop')
plt.show()
```

# Yes there is a diffrence between suicide rates of men and women because men have a lot greater suicide number.
